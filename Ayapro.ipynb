```python
# Convert multiple columns to lowercase
columns_to_lower = ['EmployeeName', 'JobTitle', 'Agency']
df[columns_to_lower] = df[columns_to_lower].apply(lambda x: x.str.lower())
```

```python
# Remove duplicates based on multiple columns (e.g., 'EmployeeName' and 'JobTitle')
df.drop_duplicates(subset=['EmployeeName', 'JobTitle'], inplace=True)
```

```python
# Specify columns for which you want to replace empty cells
columns_to_fill = ['TotalPay', 'BasePay', 'OvertimePay','OtherPay']

# Calculate the median for the specified columns
median_values_subset = df[columns_to_fill].median()

# Replace empty cells in the specified columns with their respective median values
df[columns_to_fill] = df[columns_to_fill].fillna(median_values_subset)

```

```python
# Remove multiple columns
columns_to_remove = ['Benefits','Notes', 'Status']
df.drop(columns=columns_to_remove, inplace=True)
```

```python

# Round values in multiple columns and then convert to integers
columns_to_round = ['TotalPay', 'BasePay', 'OvertimePay']
df[columns_to_round] = df[columns_to_round].round().astype(int)

```
